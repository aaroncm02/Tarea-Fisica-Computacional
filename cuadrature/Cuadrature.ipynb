```python
import numpy as np
import matplotlib.pyplot as plt

# Función que entrega puntos y pesos para N usando polinomios de Legendre
def gaussxw(N):
    x, w = np.polynomial.legendre.leggauss(N)
    return x, w

# Función que escala el intervalo de integración de [-1,1] a [a,b]
def gaussxwab(a, b, x, w):
    return 0.5 * (b - a) * x + 0.5 * (b + a), 0.5 * (b - a) * w

# Definición del integrando
def integrando(x):
    return np.sin(x**2)

# Límites de integración
lim_inf = 0.0
lim_sup = np.pi

# Inicializar listas para guardar resultados
Ns = list(range(2, 51))  # N desde 2 hasta 50
resultados = []

# Calcular integral para cada N
for N in Ns:
    x_sin_escalar, w_sin_escalar = gaussxw(N)
    x_escalado, w_escalado = gaussxwab(lim_inf, lim_sup, x_sin_escalar, w_sin_escalar)
    resultado = np.sum(w_escalado * integrando(x_escalado))
    resultados.append(resultado)

# Graficar resultados
plt.plot(Ns, resultados, 'o-', label=r'$\int_0^\pi \sin(x^2)\,dx$')
plt.xlabel('N')
plt.ylabel('Resultado de la integral')
plt.title(r'Cuadratura Gaussiana para $\int_0^\pi \sin(x^2)\,dx$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

#Comprobar 
print(f"Valor estimado de la integral con N=50: {resultados[-1]:.5f}")
```
